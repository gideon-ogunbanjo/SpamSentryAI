```python
# here i renamed the v2 column to EmailText and the v1 column to Label. This will make it easier to distinguish while training
data.rename(columns = {'v2':'EmailText'}, inplace = True)
data.rename(columns = {'v1':'Label'}, inplace = True)
```

## Training the model

```python
# importing libraries
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import svm 
```

```python
X = data['EmailText']
y = data["Label"]
z_train, z_test,y_train, y_test = train_test_split(X,y,test_size = 0.2)
```

```python
# Feature Extraction
cv = CountVectorizer()
features = cv.fit_transform(z_train)
```

```python
print(features)
```

```python
# building the model using SVM
Sentry = svm.SVC()
Sentry.fit(features,y_train)
```

Evaluating the Model

```python
from sklearn.metrics import accuracy_score

features_test = cv.transform(z_test)
predictions = Sentry.predict(features_test)
accuracy = accuracy_score(y_test, predictions)
print("Model Accuracy: {:.2f}%".format(accuracy * 100))
```
